```python
x = 10  
precisao = 1E-06  
iteracoes = 0  

while x > precisao:  
    x = x / 2  
    iteracoes += 1 

print("Último valor de x: ", x)
print("Número de iterações: ", iteracoes)
```

```python
import matplotlib.pyplot as plt

x = 10
precisao = 1E-06
iteracoes = 0
valores_x = []  
iteracoes_y = []  

while x > precisao:
    valores_x.append(x)
    iteracoes_y.append(iteracoes)
    x = x / 2
    iteracoes += 1

plt.plot(iteracoes_y, valores_x, '-o')
plt.xlabel('Número de iterações')
plt.ylabel('Valor de x')
plt.show()
```
